```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Read & Preview the dataset
file_path = '/Users/sandip/code/GreyAtom/greyatom_Jupyter_notes/DataFiles/datasets/automobile_data.csv'

df = pd.read_csv(file_path, header=0) 
```

```python
# check datatypes of all fields
df.info()
```

```python
# check descriptive stats
df.describe()
```

```python
# find null fields
```

```python
# Which manufacturer/make makes highest number of cars
df['make'].value_counts().plot(kind='bar')
plt.show()
```

```python
# which is more preferred fuel type
df['fuel-system'].value_counts().plot(kind='bar')
plt.show()
```

```python
# correlation b/w engine size and price
df['price'] = df['price'].replace(to_replace='?', value=0).astype(float).replace(to_replace=0, \
                                                                                 value=df['price'].mean())
plt.scatter(df['engine-size'], df['price'])
plt.show()


# Code for linear regression
import seaborn as sns
sns.regplot(df['engine-size'], df['price'])
plt.show()
```

```python
# Make a box of price and make to answer following
# Most expensive car manufacturer, manufacturer for afforabled/less expensive cars
plt.rcParams['figure.figsize'] = (23,10)
sns.boxplot(x='make', y='price', data=df)
plt.show()
```

```python
# How is the curb weight related to highway mpg and city mpg for diffrent car makers
sns.lmplot('city-mpg', 'curb-weight', df, hue='make', fit_reg=False, size=8, legend_out=True)
plt.show()
```
